```python
marks = int(input("Enter Marks:"))
if marks<0 or marks > 100:
    print("marks invalid")
elif 80<=marks<=100:
    print("A grade")
elif 60<=marks<=79:
    print("B grade")
elif 41<=marks<=59:
    print("C grade")
elif 30<=marks<=40:
    print("D grade")
else:
    print("F grade")
```
