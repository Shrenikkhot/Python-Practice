```python
data = "python"
print(data)
```

```python
print(data[2])
```

```python
# 1.upper() converts the string to uppercase
print(data.upper())
```

```python
# 2.lower() converts the string to lowercase
print(data.lower())
```

```python
data = "codes with python"
#3. partition() returns a tuple
print(data.partition("w"))
```

```python
#4. replace() replaces substring inside
print(data.replace("python","shrenik"))
```

```python
#5. find() returns the index of first occurence of substring
print(data.find("p"))
```

```python
#6.rstrip() removes trailing characters
data = "welcome    "
print(data,"hello")
print(data.rstrip(),"hello")
```

```python
#7. split() splits string from left
data = "python r go julia java ruby"
print(data.split(" "))
```

```python
#8. startswith() checks if string starts with the specified string
print(data.startswith("codes"))
```

```python
#9. isnumeric() checks numeric characters 
number = "89F"
print(number.isnumeric())
```

```python
# 10.index() returns index of substring
print(data.index("t"))
```

```python
# 11. format() print data
name = "Shrenik"
age = 23
height = 5.11
print("your name is {0}, height{2} and age {1}".format(name,age,height))
print(f'you name is {name}, height is {height} and age is {age}')
```
