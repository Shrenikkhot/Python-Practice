```python
find_data = input('enter your word - ')
data = "Python is best"

if find_data in data:
    print(f"{find_data}:- is found....")
else: 
    print(f"{find_data}:- not found....")
```

```python
import re
data = "python is best coding shrenik is best coder of the world"

find_text = r'best'
#match()
match = re.search(find_text,data)

if match:
    print('Data found...',match.group())
else:
    print('Data not found...')
```

```python
#findall()
find_data = re.findall(find_text,data)
print(find_data)
print(len(find_data))
```
