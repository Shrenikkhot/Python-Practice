```python
class Health():
    def health_info():
        print("Health is good")
```

```python
class Food(Health):
    def food_info():
        print("Calorie intake is enough")
```

```python
class Exercise(Health):
    def exercise_info():
        print("Exercise is complete")
```

```python
health_chart = Exercise
health_chart.exercise_info()
health_chart.health_info()
```

```python
health_chart1 = Food
health_chart1.food_info()
health_chart1.health_info()
```
