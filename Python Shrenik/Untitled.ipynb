```python
#https://quotes.toscrape.com
import requests
from bs4 import BeautifulSoup
web  = "https://quotes.toscrape.com"
page = requests.get(web) #get data from website
soup = BeautifulSoup(page.text, "html.parser")
all_quotes = soup.find_all("span", class_="text")

#print(all_quotes)
for i in range(10):
    quote = all_quotes[i].text
    print(f'{i+1} - {quote}')
```

```python

all_author = soup.find_all("small", class_="author")

#print(all_quotes)
for i in range(10):
    author = all_author[i].text
    print(f'{i+1} - {author}')
```

```python
quote_div = soup.find("div",class_="quote")

if quote_div:
    quote_text = quote_div.find("span",class_="text").text
    author_text = quote_div.find("small",class_="author").text
    print(f"quote = {quote_text}")
    print(f"author = {author_text}")
    
```

```python
quote_div = soup.find_all("div",class_="quote")

for i in quote_div:
    quote_text = i.find("span",class_="text").text
    author_text = i.find("small",class_="author").text
    print(f"quote = {quote_text}|author = {author_text}")
```
