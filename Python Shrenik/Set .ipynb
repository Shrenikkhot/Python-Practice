```python
data = [1,2,3,4,5,6,7,12,1,21,21,2,1,22,2,2,2,2,22,2,3,4,5,6]
data_set = set(data)
print(data_set)
```

```python
#add
data_set.add(898)
print(data_set)
```

```python
#update
data_set.update([222,45,49])
print(data_set)
```

```python
data = list(data_set)
print(data)
```

```python
print(data[5:8])
print(data[7])
```

```python
#discard
data = {12,1,14,15,13,12,1,2}
data.discard(1)
print(data)
```

```python
data = {12,1,14,15,13,12,1,2}
data.remove(12)
print(data)
```

```python
#pop
data.pop()
print(data)
```

```python
#clear 
data.clear()
print(data)
```
