```python
class city:
    def city_info():
        print("Mumbai")
    
```

```python
class credit_card:
    def card_info():
        print("6846579845")
```

```python
class HDFCBank(city,credit_card):
    def amount_info():
        print("100000000000")
```

```python
class Client(HDFCBank):
    def Client_info():
        print("Shrenik")
    
```

```python
bank_client = Client
bank_client.amount_info()
bank_client.Client_info()
bank_client.city_info()
bank_client.card_info()
```
