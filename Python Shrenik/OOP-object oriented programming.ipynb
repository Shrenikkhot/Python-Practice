```python
class Health:
    def __init__(a):
        fees = 15000
        a.fees = fees
        print(f"Welcome to  K.F. Gyms, our annual gym membership is {fees}")
        print("This gym is located in PAN India")
    def fitness(a,calories):
        a.calories = calories
        print(f"Total calories burned: {calories}")
        print("Strength training goal achieved")
        print("Running goal achieved")
    def food(a,protein_intake):
        a.protein_intake = protein_intake
        print(f"Today's protein intake was: {protein_intake} grams")
        print("3 meals planned for the day")
        print("Protein goal achieved")
    
```

```python
health_tracker = Health()
health_tracker.fitness(1500)
health_tracker.food(200)
```
