```python
#EXAMPLE
def bill(price, gst, quantity):
    print('------------TotalBill---------------')
    print(f'Price = ₹{price}')
    print(f'GST = {gst}%')
    print(f'Quantity = {quantity}')
    total_price = price*quantity
    print(f'Total_price = ₹{total_price}')
    gst_amount = total_price*(gst/100)
    print(f'GST amount = ₹{gst_amount}')
    total = total_price + gst_amount 
    print(f'Total inclusive GST = ₹{total}')
    print('-------------------------------------')
```

```python
bill(2000,18,100)
bill(300,5,5)
bill(100,10,10)
```
