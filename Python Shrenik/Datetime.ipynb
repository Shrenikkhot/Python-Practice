```python
import datetime as dt
print(dt.datetime.today())
print(dt.datetime.now())
```

```python
from datetime import date
print(date.today())
print(date.today().year)
print(date.today().day)
print(date.today().month)
```

```python
d = dt.datetime.now().time()
print(d)
print(d.hour)
print(d.minute)
print(d.second)
print(d.microsecond*1000)
```

```python
#create a date object 
date_obj = dt.date(2024,8,5)
print("Date Object:", date_obj)
```

```python
#create a time object
time_obj = dt.time(14,30,45)
print("Time Object: ", time_obj)
```

```python
datetime_obj = dt.datetime(2024,8,5,14,30,45)
print("Datetime object: ",datetime_obj)
```
