```python
import requests 
from bs4 import BeautifulSoup
# url = "https://www.muscleblaze.com/"
url = "https://www.avvatarindia.com/"
res = requests.get(url)

soup  = BeautifulSoup(res.text, "html.parser")

title = soup.title.text
print(title)
heading = soup.find_all("a")

for i in heading:
    print(i.text)
```
