```python
import pandas as pd
#create a sample data
data = {'store':['North','South','East','East','North','North','South','South','South','North'],
'product':['laptop','phone','tablet','phone','tablet','phone','tablet','laptop','laptop','tablet'],
'sales':[1200,3400,2000,1500,6000,200,600,800,1204,1800],
'qt':[2,3,4,2,4,6,7,2,2,8]
}
#DataFrame
df =pd.DataFrame(data)
df
#df.to_csv("sales.csv") export to csv
df.to_csv("sales.csv")
```

```python
#group by one column-> total sales per store
store_sales = df.groupby('store').sum()
print(store_sales)
```

```python
#group by one column -> total sales per product
product_sales = df.groupby('product').sum()
print(product_sales)
```

```python
sales_by_store = df.groupby('store')['sales'].sum()
print(sales_by_store)
```

```python
sales_by_product =df.groupby('product')['sales'].sum()
print(sales_by_product)
```

```python
quantity_by_product =df.groupby('product')['qt'].sum()
print(quantity_by_product)
```

```python
quantity_by_store = df.groupby('store')['qt'].sum()
print(quantity_by_store)
```

```python
#agg()
summary = df.groupby('store').agg(
    {
        'sales':['sum','mean','count'],
        'qt':['sum']

    }
)
print(summary)
```

```python
#agg()
summary = df.groupby('store').agg(
    {
        'sales':['sum','mean','count'],
        'qt':['sum']

    }
)
print(summary)
```

```python
#group by using multiple column
by_store_product = df.groupby(['store','product'])['sales'].sum()
print(by_store_product)
```

```python
#unstack()
print(by_store_product.unstack(fill_value=0))
```

```python
def sales_value(group):
    group['High_value'] = group['sales']>1500
    return group
#apply 
result = df.groupby('store').apply(sales_value)
print(result[['store','sales','High_value']])
```

```python
def sales_value(group):
    group['High_value'] = group['sales']>1500
    return group
#apply 
result = df.groupby('product').apply(sales_value)
print(result[['store','sales','High_value']])
```

```python
import pandas as pd

# Sample sales data
data = {
    'Order_ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Date': ['2024-01-05', '2024-01-10', '2024-01-15', '2024-02-01', '2024-02-05',
             '2024-02-10', '2024-03-01', '2024-03-05', '2024-03-10', '2024-01-20'],
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Laptop', 'Phone', 
                'Tablet', 'Mouse', 'Keyboard', 'Phone', 'Laptop'],
    'Category': ['Electronics', 'Accessories', 'Accessories', 'Electronics', 'Electronics',
                 'Electronics', 'Accessories', 'Accessories', 'Electronics', 'Electronics'],
    'Sales': [75000, 1200, 2500, 80000, 45000, 30000, 1500, 2800, 42000, 72000],
    'Quantity': [1, 5, 3, 1, 2, 3, 4, 2, 1, 2],
    'City': ['Mumbai', 'Delhi', 'Bangalore', 'Mumbai', 'Delhi', 
             'Bangalore', 'Mumbai', 'Delhi', 'Bangalore', 'Mumbai']
}

df_temp = pd.DataFrame(data)
df_temp.to_csv('sales_data.csv', index=False)
df_temp 
```

```python

# Question 1 - Total sales by city
result1 = df_temp.groupby('City')['Sales'].sum()
print(result1)
```

```python
# Question 2 - Average quantity by category 
result2 = df_temp.groupby('Category')['Quantity'].mean().round(2)
print(result2)
```

```python
# Question 3 - (Sales by city and Category)Group by two columns (city+Category)
result3 = df_temp.groupby(['City','Category'])['Sales'].sum()
print(result3)
```

```python
#Question 4 - Count number of order per city
result4 = df_temp.groupby('City')['Order_ID'].count()
print(result4)
result4 = df_temp.groupby('City').size()
print(result4)
result4 = df_temp.groupby('City').count()
print(result4)
```

```python
#Question 5 - Sales ->sum,mean,max(city) 
#               Quantity ->sum
result5 = df_temp.groupby('City').agg(
    {
        'Sales':['sum','mean','max'],
        'Quantity': ['sum']
    }
).round(2)
print(result5)
```
